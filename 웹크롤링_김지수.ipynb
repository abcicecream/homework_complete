```python
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
import requests
```

## 멜론 크롤링

```python
url = 'https://www.melon.com/chart/index.htm'
```

```python
ua = UserAgent()
```

```python
# 헤더 선언
headers = {
'User-Agent': ua.chrome,
'referer': 'https://naver.com'
}
```

```python
response = requests.get(url, headers=headers)
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')   # 링크 나와서 안되면 커널 끊었다가 다시 하면 됨
print(soup.prettify())
```

```python
titles = soup.find_all('div',{'class':'ellipsis rank01'})
artists = soup.select('span.checkEllipsis')
```

```python
musics = []
names = []

for title, artist in zip(titles, artists):
    musics.append(title.get_text().strip()), names.append(artist.get_text().strip())
```

```python
import pandas as pd

melon = pd.DataFrame({'Service':'Melon','title':musics,'artist':names})
melon.index = [f'{i}위' for i in range(1,101)]
melon.index.name = '순위'
```

```python
melon
```

```python
melon.to_excel('melonchart.xlsx')
```

## 벅스 크롤링

```python
url = 'https://music.bugs.co.kr/chart'

html = urlopen(url)
html
```

```python
soup = BeautifulSoup(html, 'lxml')
soup
```

```python
print(soup.prettify())
```

```python
# 실험
a = soup.select('a.artistTitle')
a
```

```python
soup.select('p.artist > a:first-child')
```

```python
titles = soup.select('p.title')
artists = soup.select('p.artist > a:first-child')

# for titles, artists in zip(t_titles, t_artists):
#     titles = titles.strip()
#     artists = artists.strip()
music = []
name = []

for title, artist in zip(titles, artists):
    music.append(title.text.strip())
    name.append(artist.text.strip())
# import pandas as pd
len(music)

import pandas as pd
bugs = pd.DataFrame({'Service':'Bugs',"title": music,
                    'artist': name})
bugs.index = [f'{i}위' for i in range(1,101)]

bugs.index.name = '순위'
```

```python
bugs
```

```python
bugs.to_excel('bugschart.xlsx')
```

## 지니 크롤링

```python
url = 'https://www.genie.co.kr/chart/top200'
```

```python
ua = UserAgent()
```

```python
# 헤더 선언
headers = {
'User-Agent': ua.chrome,
'referer': 'https://naver.com'
}
```

```python
response = requests.get(url, headers=headers)
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```

```python
titles = soup.select('a.title.ellipsis')
artists = soup.select('td.info a.artist.ellipsis')
```

```python
# 확인
artists
```

```python
musics = []
names = []

for title, artist in zip(titles, artists):
    musics.append(title.get_text().strip())
    names.append(artist.text.strip())
```

```python
import pandas as pd

genie = pd.DataFrame({'Service':'Genie','title':musics,
                      'artist':names})

genie.index = [f'{i}위' for i in range(1,51)]

genie.index.name = '순위'

genie
```

```python
genie.to_excel('geniechart.xlsx')
```

## 병합

```python
import pandas as pd
```

```python
mel = pd.read_excel('melonchart.xlsx')
mel
```

```python
gen = pd.read_excel('geniechart.xlsx')
gen.drop(columns=['순위'],inplace=True)
```

```python
bug = pd.read_excel('bugschart.xlsx')
bug.drop(columns=['순위'],inplace=True)
```

```python
music_concat = pd.concat([mel,bug,gen],join='outer',axis=1)
music_concat
```

```python
music_concat.set_index('순위',inplace=True)
```

```python
music_concat.to_excel('total.xlsx')
```
